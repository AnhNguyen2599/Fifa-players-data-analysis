# 4. Show number of players and their countries.

```python
#To show number of players in each countries - option 1
#we can use groupby(), count(), then rename()
mdf = df[["sofifa_id","nationality"]].groupby('nationality').count()
mdf = mdf.rename(columns={"sofifa_id":"number_of_players"})
mdf
```

```python
#To show number of players in each countries - option 2
#Use .agg(new_col_name=('col_name', 'agg_func')
mdf2 = df.groupby('nationality').agg(number_of_players=('sofifa_id','count'))
mdf2
```

## 5. If you find many records in point 4 then show only top 10 countries and their number of players.

```python
mdf.sort_values(by=['number_of_players'],ascending=False).head(10)
```

# 6.Now you have to create a bar plot of top 5 countries and their number of players, try to fill green color in bars.

```python
#first, need to import matplotlib.pyplot (done in cell 1)
mdf3 = mdf.sort_values(by='number_of_players', ascending=False).head(5)
plt.figure(figsize=(8,8))
plt.bar(mdf3.index,mdf3["number_of_players"],color='green')
plt.xlabel("Nationality")
plt.ylabel("Number of players")
plt.title("Top 5 countries with highest number of players")
plt.annotate
for x,y in zip(mdf3.index,mdf3["number_of_players"]): 
    label = y
    plt.annotate(label,
                 (x,y),                         #coordinates to position the label
                 textcoords="offset points",    #how to position the label
                 xytext=(0,3),                  #distance from label to (x,y)
                 ha="center")                   #how to align the label

plt.show()
```
